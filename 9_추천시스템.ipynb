# 잠재 요인 협업 필터링 

## 잠재 요인 협업 필터링의 이해 

: 숨겨져 있는 '잠재 요인'을 기반으로 분해된 매트릭스를 이용해 사용자가 아직 평가하지 않은 아이템에 대한 예측 평가를 수행하는 것
1. 사용자-아이템 행렬 데이터(다차원 희소행렬) -> 사용자-잠재 요인 행렬, 아이템-잠재 요인 행렬(저차원 밀집행렬) 로 분해 
2. 분해된 두 행렬의 내적 => 예측 사용자-아이템 행렬 데이터 생성 => 평점을 부여하지 않은 아이템에 대한 예측 평점 생성!

## 확률적 경사 하강법을 이용한 행렬 분해 

PROBLEM : SVD는 null값이 없는 행렬에만 적용할 수 있음<BR>
SOLUTION : 확률적 경사 하강법이나 ALS 방식을 이용해 SVD를 수행 <BR>
**분해된 행렬들로 계산된 예측 사용자-아이템 행렬 값이 실제 사용자-아이템 행렬값과 가장 최소의 오류를 가질 수 있도록 <BR>반복적인 비용 함수 최적화를 통해 분해된 행렬 유추** 

```python
import numpy as np

R = np.array([[4, np.NaN, np.NaN, 2, np.NaN ],
              [np.NaN, 5, np.NaN, 3, 1 ],
              [np.NaN, np.NaN, 3, 4, 4 ],
              [5, 2, 1, 2, np.NaN ]]) # 원본 사용자-아이템 행렬
num_users, num_items = R.shape
K = 3

np.random.seed(1)
P = np.random.normal(scale = 1./K, size = (num_users, K)) # 사용자-잠재요인 행렬 랜덤
Q = np.random.normal(scale = 1./K, size = (num_items, K)) # 아이템-잠재요인 행렬 랜덤 
```

```python
from sklearn.metrics import mean_squared_error

def get_rmse(R, P, Q, non_zeros):
  error = 0
  full_pred_matrix = np.dot(P, Q.T) # 예측 사용자-아이템 행렬 

  x_non_zero_ind = [non_zero[0] for non_zero in non_zeros]
  y_non_zero_ind = [non_zero[1] for non_zero in non_zeros]
  R_non_zeros = R[x_non_zero_ind, y_non_zero_ind] # 실제 사용자-아이템 평점
  full_pred_matrix_non_zeros = full_pred_matrix[x_non_zero_ind, y_non_zero_ind] # 예측 사용자-아이템 평점

  mse = mean_squared_error(R_non_zeros, full_pred_matrix_non_zeros)
  rmse = np.sqrt(mse)

  return rmse
```

```python
non_zeros = [(i, j, R[i, j]) for i in range(num_users) for j in range(num_items) if  R[i, j]>0]

steps = 1000 # 업데이트 횟수
learning_rate = 0.01 # 학습률
r_lambda = 0.01 # 규제 강도

for step in range(steps):
  for i, j, r in non_zeros:
    eij = r-np.dot(P[i,:], Q[j,:].T) # 예측 평점

    P[i, :] = P[i, :]+learning_rate*(eij*Q[j,:]-r_lambda*P[i, :])
    Q[j, :] = Q[j, :]+learning_rate*(eij*P[i,:]-r_lambda*Q[j, :])
  rmse = get_rmse(R, P, Q, non_zeros)
  if(step%50)==0:
    print("### iteration step : ", step, " rmse : ", rmse)
```

```python
pred_matrix = np.dot(P, Q.T)
print(np.round(pred_matrix,3)) # 예측 행렬 
```
